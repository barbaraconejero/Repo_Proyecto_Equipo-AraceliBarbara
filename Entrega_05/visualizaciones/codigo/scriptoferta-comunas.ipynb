```python
# Limpiar texto: eliminar espacios, mayúsculas y caracteres extra
df['Termómetro Vegano'] = df['Termómetro Vegano'].str.strip().str.lower()

# Corregir manualmente las categorías parecidas
df['Termómetro Vegano'] = df['Termómetro Vegano'].replace({
    'apto para vegetarianos y opciones veganas': 'Apto para vegetarianos y opciones veganas',
    'apto para vegetarianos, opciones veganas y opciones sin gluten': 'Apto para vegetarianos, opciones veganas y opciones sin gluten',
    'opciones veganas y opciones sin gluten': 'Opciones veganas y opciones sin gluten',
    'apto para vegetarianos': 'Apto para vegetarianos',
    'opciones veganas': 'Opciones veganas'
})

# Volver a contar
conteo_limpio = df['Termómetro Vegano'].value_counts().reset_index()
conteo_limpio.columns = ['Categoría', 'Cantidad']

conteo_limpio
```

```python
# Función para clasificar automáticamente
def clasificar_restaurante(valor):
    if pd.isna(valor):
        return "Desconocido"
    elif "opciones" in valor or "vegetarianos" in valor:
        # Si menciona opciones o vegetarianos, lo consideramos tradicional adaptado
        return "Tradicional con opciones veganas"
    else:
        # Si no menciona esas palabras, lo asumimos como vegano 100 %
        return "Vegano 100%"

# Aplicar la función a cada fila
df['Tipo de restaurante'] = df['Termómetro Vegano'].apply(clasificar_restaurante)

# Revisar si se creó correctamente
df[['Termómetro Vegano', 'Tipo de restaurante']].head(10)
```

```python
def clasificar_restaurante(valor):
    if pd.isna(valor):
        return "Desconocido"
    valor = valor.lower().strip()
    if "opciones" in valor or "vegetarianos" in valor:
        return "Tradicional con opciones veganas"
    elif "vegano" in valor and "opciones" not in valor and "vegetarianos" not in valor:
        return "Vegano 100%"
    else:
        return "Desconocido"

# Aplicar la función
df['Tipo de restaurante'] = df['Termómetro Vegano'].apply(clasificar_restaurante)

# Revisar cuántos hay en cada tipo
df['Tipo de restaurante'].value_counts()
```

```python
import altair as alt

grafico = alt.Chart(conteo_comuna).mark_bar(color='#74c69d').encode(
    x=alt.X('Cantidad:Q', title='Cantidad de restaurantes veganos/vegetarianos'),
    y=alt.Y('Comuna:N', sort='-x', title='Comuna'),
    tooltip=['Comuna', 'Cantidad']
).properties(
    title='Concentración de restaurantes veganos por comuna en Santiago',
    width=700,
    height=400
)

grafico
```

```python
grafico = (
    alt.Chart(conteo_comuna)
    .mark_bar(cornerRadiusTopLeft=5, cornerRadiusTopRight=5, color="#74c69d")
    .encode(
        x=alt.X("Cantidad:Q", title="Cantidad de restaurantes veganos/vegetarianos"),
        y=alt.Y("Comuna:N", sort="-x", title="Comuna"),
        tooltip=["Comuna", "Cantidad"]
    )
    .properties(
        title="Concentración de restaurantes veganos por comuna en Santiago",
        width=700,
        height=400
    )
    .configure_title(
        fontSize=16,
        font="Arial",
        anchor="start",
        color="#333"
    )
)

grafico
```
